**Import Library**

```python
import pandas as pd 
```

```python
set = pd.read_csv('../Data/top20_set_prices.csv')
```

**Data cleaning for SET and S&P500**

**SET**

```python
set.shape
```

```python
set
```

```python
# drop all the columns except for Adj Close
set = set[[col for col in set.columns if col.startswith(('Price', 'Adj'))]]
```

```python
set.columns
```

```python
# Reset the index by the name of the stocks 
set.columns = set.iloc[0]
# Drop all the stocks name rows and NaN rows 
set = set[2:]
```

```python
set.isnull().sum()[set.isnull().sum() != 0]
```

```python
# Seperate the (CRC, GULF, OR, and SCB) into different 
the_rest = set[['CRC.BK', 'GULF.BK', 'OR.BK','SCB.BK']]
the_rest
```

```python
main = set.drop(columns = ['CRC.BK', 'GULF.BK', 'OR.BK','SCB.BK'])
main = main.rename(columns ={'Ticker':'Date'})
```

```python
# set date to date 
main['Date'] = pd.to_datetime(main['Date'])

# set date into index 
main = main.set_index('Date')
```

```python
# Converting all the obj into float
for col in main.columns[1:21] :
        main[col] = main[col].astype(float)
```

```python
# Export the file 
main.to_csv("../Data/SET_main.csv")
the_rest.to_csv("../Data/SET_residual.csv") 
```
